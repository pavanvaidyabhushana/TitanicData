```python
def male_female_child(object):
    age,sex = object
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
data['person'] = data[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
deck = data['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    
cabindf = pd.DataFrame(levels,columns=['Cabintype'])
sns.factorplot(kind='count',x='Cabintype',data=cabindf,palette='winter_d')
```

```python
cabindf = cabindf[cabindf.Cabintype != 'T']
sns.factorplot(x='Cabintype',kind='count',data=cabindf,palette='summer')
```

```python
sns.factorplot(x="Embarked",data= data,hue='Pclass',
               kind='count',palette='winter_d',
               order=['C','Q','S'])
```

```python
data["Agegroup"] = pd.qcut(data["Age"],4)
sns.barplot(data = data , y = "Survived" , x = "Agegroup").set(title = "Survived wrt Age groups")
```

```python
sns.catplot(data = data , x = "Pclass" , col= "Embarked" ,kind = "count" ,hue = "Survived")
```

```python
sns.catplot(data = data , col = "Embarked" , hue = "Sex" , kind = "count" , x = "Survived").set(
    title = "No of male/female survived at Embarked position")
```

```python
Family = data["Parch"] + data["SibSp"]
data = data.assign(Family_size = Family)
data
```

```python
sns.barplot(data= data , x = "Family_size" , y = "Survived").set(
title = "% of ppl survived in each family depends on size")
```

```python
data.loc[data['Family_size'] == 0,'FamilySize'] = "Alone"
data.loc[data['Family_size'] != 0,'FamilySize'] = "Family"
sns.catplot(data = data , x = "FamilySize" , kind = "count" , hue = "Survived").set(
    title = "Survived ")
```

```python
sns.factorplot(x='FamilySize',data = data,kind='count',palette='Blues')
```

```python
data['Survival'] = data.Survived.map({0:'NO',1:'YES'})
sns.factorplot(x='Survival',data=data,kind='count',palette='Set1')
```
